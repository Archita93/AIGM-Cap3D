```python
import pandas as pd

baseline = pd.read_csv("eval_results_finetuned.csv")
finetuned = pd.read_csv("eval_results.csv")

# Get averages rows
avg_base = baseline[baseline["uid"] == "AVERAGE"].iloc[0]
avg_fine = finetuned[finetuned["uid"] == "AVERAGE"].iloc[0]

metrics = {
    "Chamfer Distance":  ("chamfer",          "lower"),
    "CLIP Score":        ("clip_score",        "higher"),
    "CLIP Similarity":   ("clip_similarity",   "higher"),
    "F-Score":           ("f_score",           "higher"),
    "R-Precision":       ("clip_r_precision",  "higher"),
    "LPIPS":             ("lpips",             "lower"),
}
```

```python
print(f"\n{'Metric':<22} {'Baseline':>10} {'Fine-tuned':>12} {'Delta':>10} {'Better?':>10}")
print("─" * 68)

for label, (col, direction) in metrics.items():
    b = float(avg_base[col])
    f = float(avg_fine[col])
    delta = f - b
    if direction == "higher":
        better = "fine-tuned" if delta > 0 else "baseline"
    else:
        better = "fine-tuned" if delta < 0 else "baseline"
    print(f"{label:<22} {b:>10.4f} {f:>12.4f} {delta:>+10.4f} {better:>10}")

print("─" * 68)
```
